```python
#Añadir al dafaframe la longuitud y latitud ws84 y epgs25830
longui = []
lati = []
lon_25830 = []
lat_25830 = []

#Coordendas WS84
coord = df['gps_coordinates']

for n in range(len(df.index)):
  c = coord[n].split(",")
  lon, lat = (c[0][13:23]), (c[1][14:23])
  longui.append(lon)
  lati.append(lat)
  
  #tranformamos a 25830
  transformer = Transformer.from_crs("epsg:4326", "epsg:25830")
  x2,y2 = transformer.transform(lon,lat)
  lon_25830.append(x2)
  lat_25830.append(y2)
  
#Listas a columnas de dataframe
df['lon_ws84'] = longui
df['lat_ws84'] = lati
df['lon_25830'] = lon_25830
df['lat_25830'] = lat_25830

df.head(3)
```

```python
#geolocalizar coordendas WS84 en cartociudad
from urllib.request import urlopen
from bs4 import BeautifulSoup
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
#Extraer datos de las coordenadas en cartociudad. Importante: referencia catastral
for z in range(len(df.index)):
  c = coord[z].split(",")
  #lon,lat=float(c[0][13:23]),float(c[1][13:24])
  lon, lat = (c[0][13:23]), (c[1][14:23])
  #print(lon,'---',lat)
  quote_page = 'https://www.cartociudad.es/geocoder/api/geocoder/reverseGeocode?lon='+lat+'&lat='+lon+'&type=refcatastral'
  page = urlopen(quote_page)
  soup = BeautifulSoup(page, 'html.parser')
  if len(soup) == 0:
    print("NO LOCALIZADO")
  else:
    print(soup)
```

```python
#Extraer y listar referencias catastrales y enlaces fotografia e informacion parcela en  catastro
referencias = []
web_foto = []
info_catastro = []

for z in range(len(df.index)):
  c = coord[z].split(",")
  #lon,lat=float(c[0][13:23]),float(c[1][13:24])
  lon, lat = (c[0][13:23]), (c[1][14:23])
  #print(lon,'---',lat)
  quote_page = 'https://www.cartociudad.es/geocoder/api/geocoder/reverseGeocode?lon='+lat+'&lat='+lon+'&type=refcatastral'
  page = urlopen(quote_page)
  soup = BeautifulSoup(page, 'html.parser')
  texto = str(soup)
  pos = texto.rfind('address')
  
  if len(soup) == 0:
    #print("NO LOCALIZADO")
    ref_cat = '00000000000000'
  else:
    ref_cat = texto[pos+10:pos+24]
    #print("Referencia catastral: ",ref_cat)
    #print(soup)
  web = 'http://ovc.catastro.meh.es/OVCServWeb/OVCWcfLibres/OVCFotoFachada.svc/RecuperarFotoFachadaGet?ReferenciaCatastral='+ref_cat
  web_cat = "https://www1.sedecatastro.gob.es/CYCBienInmueble/OVCListaBienes.aspx?del=3&rc1="+ref_cat[0:7]+"&rc2="+ref_cat[7:14]+"'"
  info_catastro.append(web_cat)
  referencias.append(ref_cat)
  web_foto.append(web)
```

```python
df['foto_catastral'] = web_foto
df['ref_cat'] = referencias
df['info_catastro'] = info_catastro
df.head(3)
```
